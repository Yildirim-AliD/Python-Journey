```python
# Develop an advanced calculator using built-in functions from the math module.

import math

def show_menu():
    print("\n--- Advanced Calculator ---")
    print("1. Square Root")
    print("2. Power")
    print("3. Logarithm")
    print("4. Sine")
    print("5. Cosine")
    print("6. Tangent")
    print("7. Factorial")
    print("0. Exit")

while True:
    show_menu()
    choice = input("Choose an operation (0-7): ")

    if choice == '0':
        print("Exiting...")
        break

    elif choice == '1':
        num = float(input("Enter a number: "))
        print("Result:", math.sqrt(num))

    elif choice == '2':
        base = float(input("Enter the base: "))
        exp = float(input("Enter the exponent: "))
        print("Result:", math.pow(base, exp))

    elif choice == '3':
        num = float(input("Enter a number: "))
        base = input("Enter base (press Enter for natural log): ")
        if base == "":
            print("Result:", math.log(num))
        else:
            print("Result:", math.log(num, float(base)))

    elif choice == '4':
        angle = float(input("Enter angle in degrees: "))
        print("Result:", math.sin(math.radians(angle)))

    elif choice == '5':
        angle = float(input("Enter angle in degrees: "))
        print("Result:", math.cos(math.radians(angle)))

    elif choice == '6':
        angle = float(input("Enter angle in degrees: "))
        print("Result:", math.tan(math.radians(angle)))

    elif choice == '7':
        num = int(input("Enter a non-negative integer: "))
        print("Result:", math.factorial(num))

    else:
        print("Invalid choice. Please try again.")
```
